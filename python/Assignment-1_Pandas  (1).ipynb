```python
# Q5 Which was the most ordered item? and How many items were ordered?
gdf=df.groupby('item_name')
gdf=gdf.sum()
gdf=gdf.sort_values('quantity',ascending=False)
gdf.head(1)
```

```python
# Q6 What was the most ordered item in the choice_description column?
grdf=df.groupby('choice_description')
grdf=grdf.sum()
grdf=grdf.sort_values('quantity',ascending=False)
grdf.head(1)
```

```python
#Q7 Turn the item price into a float
#df.dtypes 
def num_float(x):
    return float(x[1:-1])

df.item_price = df.item_price.apply(num_float)

df.dtypes
```

```python
# Q8 print a data frame with only two columns item_name and item_price

df1=df[['item_name','item_price']]
df1
#type(df1)
```

```python
#Q9 delete the duplicates in item_name and quantity
#df1=pd.read_table(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
#df1.head()
#df1.drop_duplicates()
df1=df.drop_duplicates(['quantity','item_name'])
df1
```
